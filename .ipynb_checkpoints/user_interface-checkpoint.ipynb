```python
import analysis_module as md
from analysis_module import extract_scene_features
scene_list = md.find_scenes(video_path)
scene_frames, scene_frames_time = md.save_scene_frames(scene_list, video_path)
key_scenes = md.find_key_scenes(scene_frames, scene_list)
seg_scene_frame_summary, seg_scene_features_summary, df_scenes_featurs_summary = md.create_features(key_scenes, video_path)
```

```python
for i in range(10):
    print("Similarity Matrix of Scenes: ")
    md.similarity_matrix(df_scenes_featurs_summary)
    print("Select 2 Scenes: ")
    print("\n")
    scene_1_ind = int(input("Scene 1 Index: ")) + 1
    scene_2_ind = int(input("Scene 2 Index: ")) + 1
    input_scene = [scene_1_ind, scene_2_ind]
    if ((1 <= scene_1_ind <= 5) and (1 <= scene_2_ind <= 5)):
        md.statistical_report(input_scene, seg_scene_frame_summary, seg_scene_features_summary, df_scenes_featurs_summary)
        print("\n")
        print("-"*100)
        ui_status = input("Continue to analyze? [yes, no]")
        if ui_status == "no":
            break
    else:
        print("Please Revise Your Index")
```
